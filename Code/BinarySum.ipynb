```python
#배열의 합을 구하기
from random import *

n=int(input("몇 개의 숫자를 생성할까요?"))
A=[]
for i in range(0,n):
  A.append(randint(1,100))
  #A.append(i+1)
print(A)

def arraySum(A):
  sum=0
  for i in range(0, len(A)):
    sum=sum+A[i]
  return sum

total_sum=arraySum(A)
print("배열의 합은", total_sum,"입니다.")
```

```python
#배열 A에 들어있는 i부터 n개의 합
import math

n=int(input("몇 개의 숫자를 생성할까요?"))
A=[]
for i in range(0,n):
  #A.append(randint(1,100))
  A.append(i+1)
print(A)
def binarySum(A, i, n):
  if n==1:
    return A[i]
  else:
    #return binarySum(A,i,int(n/2))+binarySum(A,int(i+n/2),int(n/2))
    return binarySum(A,i,math.floor(n/2))+binarySum(A,math.floor(i+n/2),math.ceil(n/2)) #math.floor:버림, math.ceil:올림

total_sum=binarySum(A,0,len(A))
print("배열의 합은", total_sum,"입니다.")

```
