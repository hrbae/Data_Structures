```python
def matMulti(A,B):
  answer=[len(B[0])*[0] for i in range(len(A))]
  if len(A[0]) ==len(B): #A의 열의 갯수=B의 행의 갯수
    for i in range(0, len(A)): #A의 행의 길이만큼
      for j in range(0, len(B[0])): #B의 열의 길이 만큼
        for k in range(0, len(A[0])): #A행렬의 열 또는 B행렬의 행
          answer[i][j] += A[i][k]*B[k][j]

  return answer

A=[[1,2],[3,4]]
B=[[1,2],[3,4]]

print(matMulti(A,B))
```

```python
def MarkovProcess(I, R, n): #n: 돌리는 횟수, 시행횟수
  ret=I
  print("Initial state", ret)
  for i in range(n):
    ret=matMulti(ret,R)
    print("At time", i+1,":", ret)

I=[[0.55,0.45]]
R=[[0.7,0.3],[0.1,0.9]]

#MarkovProcess(I,R,1)
MarkovProcess(I,R,100)
```
