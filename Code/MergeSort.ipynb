```python
from random import *

def merge_sort(S):# S는 리스트
  if (len(S) <= 1):
    return S
  else: #나누는 단계
    mid=len(S)//2 # 나누기가 두개가 있으면 나머지는 버려버린다는 뜻
    sl=S[:mid]
    sr=S[mid:]
    s1=merge_sort(sl)
    s2=merge_sort(sr)
    return merge(s1,s2)

def merge(left, right): #merge하는 단계
  result=[]
  while len(left)>0 or len(right)>0:
    if len(left)>0 and len(right)>0:
      if left[0]<=right[0]:
        result.append(left[0])
        left=left[1:]
      else:
        result.append(right[0])
        right=right[1:]
    elif len(left)>0:
      result.append(left[0])
      left=left[1:]
    elif len(right)>0:
      result.append(right[0])
      right=right[1:]
  return result

n= int(input("몇 개의 숫자를 생성할까요?"))
S=[]
for i in range(n):
  S.append(randint(1, 100))

print(S)
print(merge_sort(S))
```
