```python
def BinaryFib(k):
  if k<=1:
    return k
  else:
    return BinaryFib(k-1)+BinaryFib(k-2)

print(BinaryFib(5))
```

```python
def LinearFib(k):
  if k==1:
    return [k,0]
  else:
    i=LinearFib(k-1)[0]
    j=LinearFib(k-1)[1]
    return [i+j, i]

#print(LinearFib(5))
#print(LinearFib(5)[0])
k=5
print("Fibonacci", k, "is", LinearFib(k)[0])
```
