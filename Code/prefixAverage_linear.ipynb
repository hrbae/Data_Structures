#Prefix Average_linear
- 배열의 합을 구해놓으면, 다시 합을 구할 필요가 없다.
image.png

```python
from random import * #난수 불러오기
import time
n=int(input("몇 개의 원소를 가진 배열을 만드시겠습니까?"))
X=[]
for i in range(0, n): #0에서 부터 n-1까지의 배열 만들어주기
  X.append(i+1)
print(X)
A=list(range(0,n))
```

```python
#linear하지 않은 버전
start = time.time() #시작시간

for i in range(0,n): #i는 0부터 n-1까지 돈다. i가 0일때는 j 반복문의 범위가 (1,0)이 되므로 돌아가지 않음. i가 2부터 j는 1,2가 돌아가게 된다. s=X[0]+X[1]+X[2]
  s= X[0]
  for j in range(1,i+1): #뒤쪽에 있는거 제외이기 때문에 i아니라 i+1
    s=s+X[j]
  A[i]= s/(i+1)
print(A)
end = time.time() #종료시간
print(f"소요시간: {end - start:.5f} sec")
```

```python
start = time.time() #시작시간
s=0 #초기값 설정
for i in range(0,n): #i는 0부터 n-1까지 돈다.
  s= s+X[i] #저장된 s값에 X값을 추가해준다.
  A[i]= s/(i+1)
print(A)
end = time.time() #종료시간
print(f"소요시간: {end - start:.5f} sec")
```
