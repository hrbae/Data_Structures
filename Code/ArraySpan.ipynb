```python
#배열로 Span 만들기
def ArraySpan(X,n): #우리가 주는 X값과 n는 배열 X의 길이(사이즈)
  S=[0]*n
  for i in range(0,n):
    s=1
    while (s<=i) and (X[i-s])<=X[i]:
      s=s+1

    S[i]=s
  return S

X=[6, 3,  4, 5, 2]
R=ArraySpan(X, len(X))
print(R)  #1,1,2,3,1 출력
```

```python
#Stack으로 Span 만들기
from Stack import Stack
def ArraySapn2(X,n):
  S=[0]*n
  A=Stack(n)

  for i in range(0,n):
    while (not A.isEmpty()) and (X[A.topv()]<=X[i]):
      A.pop()
    if A.isEmpty():
      S[i]=i+1
    else:
      S[i]=i-A.topv()
    A.push(i)
  return S
X=[6, 3,  4, 5, 2]
R=ArraySapn2(X, len(X))
print(R)  #1,1,2,3,1 출력
```
