#[Prime Number Discrimination]


1.   정수 N을 입력받는다.
2.   2 ~ N-1까지의 수로 N을 나누고 나머지를 확인한다.
3. 모든 나머지가 0이 아니라면 그 수를 소수로 판단한다.



```python
no = int(input("어느 범위에서 소수를 구하고 싶은가요?"))
```

```python
print(no)
```

```python
l_no = list(range(2, no+1))
```

```python
print(l_no)
```

```python
for k in range(0, no-1):
  i = k+2 #i는 2부터 체크
  inc=2
  temp=i*inc
  while temp <= no:
    if temp in l_no:
      l_no.remove(temp)
    inc=inc+1
    temp=i*inc
```

```python
print("찾아낸 소수:", l_no)
```
