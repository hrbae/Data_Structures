```python
from LinkedLstack import LStack
```

```python
def ParenMatch(X, n):
  S= LStack()
  for i in range(0,n):
    if isOpening(X[i]):
      S.push([X[i]])
    else: #if closing
      if S.isEmpty():
        return False
      if isMatch(X[i], S.pop()):
        return False

  if S.isEmpty():
    return True
  else:
    return False

def isOpening(a):
  if (a=='(') or (a=='{') or (a=='['):
    return True
  else:
    return False

def isMatch(a,b):
  if (a==')' and b=='(') or (a=='}' and b=='{') or (a==']' and b=='['):
    return True
  else:
    return False

X1=['{','}', '[','{', '}', ']']
X2=['(',')','{']

print(ParenMatch(X1, len(X1)))
print(ParenMatch(X2, len(X2)))
```
